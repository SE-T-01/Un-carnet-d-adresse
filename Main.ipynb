```python
def AfficheOption():
    print("1 Ajouter Contact ")
    print("2 Supprimer Contact ")
    print("3 Afficher Contact ")
    print("4 Afficher Tous le contact ")
    print("5 Quitter")
```

```python
pattern = r'^[\w\.-]+@[\w\.-]+\.\w+$'
AfficheOption()
```

```python
rep=AddressBook()
while True:
    AfficheOption()
    Entree=int(input("Choisir l'opération que vous voulez :"))
    if Entree ==1:
        Nom=input("Saisez le nom")
        assert isinstance(Nom,str) and len(Nom) !=0,"Numéro Invalid"
        Tel=input("Saisiez le numéro de téléphone")
        assert isinstance(Tel,str) and Tel.isdigit() and len(Tel)>=10 and (Tel.startswith("06") or Tel.startswith("07")) ,"Numéro de téléphone invalide"
        Email=input("Saisez l'email")
        assert isinstance(Email,str) and re.match(pattern, Email) ,"Email non valid"
        c1=Contact(Nom,Tel,Email)
        rep.AddContact(c1)
    elif Entree==2:
        Nom=input("Saisez le nom")
        assert isinstance(Nom,str) and len(Nom) !=0,"Numéro Invalid"
        rep.RemoveContact(Nom)
    elif Entree==3:
        Nom=input("Saisez le nom")
        assert isinstance(Nom,str) and len(Nom) !=0,"Numéro Invalid"
        rep.desplayContact(Nom)
    elif Entree==4:
        print("Affichage de Tous contact")
        rep.afficher_fichier()
    elif Entree==5:
         print("Sortie de Système!!")
         break
    else:
        print("Choix Invalid")
```
